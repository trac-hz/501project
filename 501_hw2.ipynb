```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
import seaborn as sns
from wordcloud import WordCloud
```

```python
df=pd.read_csv('movies.csv', encoding = "ISO-8859-1")
df.head()
```

```python
#Quantitive: Budget, Gross, Runtime, Score and Votes.
#Qualitative: Country, Genre, Rating, 
# Text:      Company, Director, Name, Star and Writer.
#Date:       Released and Year.
```

```python
df.describe().T
```

```python
sns.heatmap(df.isnull(), cbar=False) 
plt.title('missing value by column', fontsize = 15)
plt.show()
```

```python
# Check if there are missing observations
df.info()
```

```python
#see quantative data statistics
for index in ['budget', 'gross', 'runtime', 'score', 'votes', 'year']:
    print(index, 'min =', df[index].min(), '&', index, 'max =', df[index].max())
```

```python
plt.subplots(figsize=(12,8))
wordcloud = WordCloud(
                          background_color='Black',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.company))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
